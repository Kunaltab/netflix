# Recommendation System

```python
df.head()
```

```python
def clean_data(x):
        return str.lower(x.replace(" ", ""))
```

```python
features=['Title','Director','Actors','Country Availability','Summary','Genre']
df=df[features]
```

```python
for feature in features:
    df[feature] = df[feature].apply(clean_data)
```

```python
def create_soup(x):
    return x['Title']+ ' ' + x['Director'] + ' ' + x['Actors'] + ' ' +x['Country Availability']+' '+ x['Summary']+' '+ x['Genre']
```

```python
df['soup'] = df.apply(create_soup, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['soup'])

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df=df.reset_index()
indices = pd.Series(df.index, index=df['Title'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df['Summary'])
```

```python
from sklearn.metrics.pairwise import linear_kernel
# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations_new(Title, cosine_sim=cosine_sim):
    Title=Title.replace(' ','').lower()
    idx = indices[Title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df['Title'].iloc[movie_indices]
```

```python
get_recommendations_new('Breaking Bad', cosine_sim2)
```
